# üî¢ Transformasi Data

Tidak semua informasi pada data diperlukan untuk mengetahui tren penjualan pizza tiap bulan. Setelah diinspeksi ternyata hanya dibutuhkan kolom `order_id` sebagai id transaksi, `date` yang merupakan tanggal transaksi dilakukan, `category` merupakan kategori pizza yang dijual restoran.Dengan beberapa kolom tersebut akan dilakukan transformasi data sebagai berikut :

<ol>
  <li>Pastikan kolom `date` bertipe data tanggal (*date* atau *datetime*)</li>
  <li>Tambahkan satu kolom yakni `month` yang merupakan nama bulan dilakukannya transaksi</li>
  <li>Pastikan kolom bulan bertipe data kategorikal agar dapat diurutkan berdasarkan urutan bulan</li>
  <li>Tambahkan kolom lain yakni `year` yang merupakan tahun dilakukannya transaksi</li>
</ol>

Jika semua perhitungan telah dilakukan tampilkan kembali data yang akan diolah

````python
# Ambil kolom yang diperlukan saja yakni order_id, date, price dan quantity
sales_data = data_pizza[['order_id', 'date', 'category']]
````

````python
# Transformasikan kolom date yang berawal dari str menjadi datetime
sales_data = sales_data.astype({
    'date' : 'datetime64[ns]'
})
````

````python
# Ekstrak nama bulan
monthname = sales_data['date'].dt.month_name()

# Tambahkan kolom month yang menyatakan bulan dilakukan transaksi
sales_data.insert(
    loc = 2,
    column = 'month',
    value = monthname
)

# Transformasikan kolom month menjadi tipe kategorikal
months = ["January", "February", "March", "April", "May", "June",
          "July", "August", "September", "October", "November", "December"]

sales_data['month'] = pd.Categorical(
    sales_data['month'],
    categories = months,
    ordered = True
)
````

````python
# Tambahkan kolom year yang menyatakan tahun dilakukan transaksi
sales_data.insert(
    loc = 3,
    column = 'year',
    value = sales_data['date'].dt.year
)
````

````python
# Tampilkan hasilnya
display(sales_data)
````

# üßÆ Agregasi Data

Lakukan proses agregasi untuk menghitung total pesanan tiap bulan yang diperoleh dengan menghitung jumlah **order_id** yang unik (tidak sekedar di count namun perlu difilter untuk order_id tunggal), lalu dikelompokkan berdasarkan bulan dan tahun. Untuk dapat melakukan hal ini gunakan sintaks `DataFrame.groupby(by = ...)` untuk mengelompokkan data berdasarkan kolom yang ditentukan lalu lanjutkan dengan method `.agg(kolom_baru = (measure, jenis_perhitungan))`


docs : <i>https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.agg.html</i>

````python
# Menghitung Total Order per Kategori Pizza
total_order_per_category = sales_data.groupby(['category', 'month', 'year'], as_index = False)\
                                     .agg(total_order = ('order_id', pd.Series.nunique))

# Memberi rank urutan untuk penjualan best seller pizza
total_order_per_category['ranking'] = total_order_per_category.groupby(['year', 'category'])['total_order'].transform(sum)\
                                                              .rank(ascending = False, method='dense')\
                                                              .astype(int)

# Urutkan data untuk re-checking process
total_order_per_category = total_order_per_category.sort_values(
      by = ['ranking', 'year', 'month'],
      ascending = True,
      ignore_index = False
)

# Tampilkan data hasil agregasi
display(total_order_per_category)
````

1.   Menonjolkan *Preattentive Attribute*
2.   Mengurangi *Cognitive Load* dengan *Decluttering*

## üë®‚Äçüîß Reparasi Data Viz - 1 : Menonjolkan Atensi
---

Atensi yang ingin ditonjolkan pada grafik performance total order tiap kategori pizza antara lain adalah :

<ul>
  <li>Memberi <i>highlight</i> dengan menebalkan garis pada kategori yang ingin ditonjolkan. Data yang igin ditonjolkan adalah kategori pizza dengan total order paling banyak (paling laris)</li>
  <li>Mengganti <i>legend</i> dengan teks sesuai kategori yang diletakkan di ujung grafik</li>
</ul>

Langkah awal yang perlu dilakukan adalah mendapatkan index data yang ingin diberi atensi dengan mengeksekusi sintaks berikut

### üë®‚Äçüîß Reparasi Data Viz - 1A : *Menebalkan Garis*

Memberi *highlight* dengan menebalkan garis pada kategori yang ingin ditonjolkan yakni kategori pizza dengan total order paling banyak (paling laris) dapat dilakukan dengan cara melakukan looping untuk mengupdate `line.width` dengan ketentuan sebagai berikut :      

<ul>
  <li>Kategori Pizza Classic karena selalu mendominasi penjualan (grafik selalu paling atas) maka dicetak paling tebal dengan ukuran 8 (opsional)</li>
  <li>Selain kategori yang telah disebutkan, dicetak dengan tebal 5</li>
</ul>

Setelah didefinisikan lalu eksekusi perintah berikut untuk mengupdate tebal (width) pada grafik garis


```
for trace, width in zip(fig.data, tebal_garis):
    if isinstance(trace, go.Scatter):
        trace.line.width = width
```

Namun untuk dapat mengeksekusi sintaks diatas perlu untuk mengimpor modul baru yang disebut Plotly Graph Objects atau `import plotly.graph_objects as go`


````python
# Import library yang dibutuhkan
import plotly.express as px
import plotly.graph_objects as go

# Buat grafik line dasar
fig = px.line(
    total_order_per_category,
    x = 'month',
    y = 'total_order',
    color = 'category'
)

# Definisikan tebal garis
tebal_garis = [8, 5, 5, 5]

# Penebalan Garis yang ingin di highlight
for trace, width in zip(fig.data, tebal_garis):
    if isinstance(trace, go.Scatter):
        trace.line.width = width

# Tampilkan Grafik
fig.show()
````

### üë®‚Äçüîß Reparasi Data Viz - 1B : *Koreksi Warna*

Beri warna yang mencolok untuk data yang ingin ditonjolkan dan beri warna yang sedikit pudar untuk data yang tidak ingin ditonjolkan. Ketentuan pewarnaan adalah sebagai berikut :     

<ul>
  <li>Kategori Pizza Classic : #D19C4B (hijau)</li>
  <li>Selain kategori yang telah disebutkan : #FAE8C5</li>
</ul>

````python
# Import library yang dibutuhkan
import plotly.express as px
import plotly.graph_objects as go

# Definisikan warna garis
pick_color = ['#D19C4B', '#FAE8C5', '#FAE8C5', '#FAE8C5']

# Buat grafik line dasar
fig = px.line(
    total_order_per_category,
    x = 'month',
    y = 'total_order',
    color = 'category',
    color_discrete_sequence = pick_color
)

# Definisikan tebal garis
tebal_garis = [8, 5, 5, 5]

# Penebalan Garis yang ingin di highlight
for trace, width in zip(fig.data, tebal_garis):
    if isinstance(trace, go.Scatter):
        trace.line.width = width

# Tampilkan Grafik
fig.show()
````

dengan membedakan gradasi warna (memberikan warna lebih gelap untuk grafik data yang ingin di *highlight* dan warna yang lebih pudar ke grafik data selainnya) akan mengarahkan audience untuk lebih fokus ke data yang ingin di *highlight*.

## üë®‚Äçüîß Reparasi Data Viz - 2 : Penambahan *Title Chart* & Mengatur Ukuran Grafik

---

*Title* (judul) dan keterangan membuat visualisasi jauh lebih mudah dipahami. Judul harus menggambarkan dengan jelas apa yang digambarkan pada grafik namun trend pada saat ini judul grafik biasanya bersifat parabola untuk lebih menarik audience membaca grafik. Namun hal tersebut bersifat opsional saja yang terpenting judul harus menggambarkan grafik yang ingin disampaikan dengan jelas. Untuk memberi judul gunakan sintaks berikut

```
fig.update_layout(
    # Mengatur panjang & lebar figure
    width = ...,
    height = ...,

    # Memberi judul pada grafik dan mengatur format fontnya
    title = dict(
        text = ... ,
        font = dict(
           ...
        )
    )
)
```

docs : <i>https://plotly.com/python/reference/layout/</i>

````python
# Import library yang dibutuhkan
import plotly.express as px
import plotly.graph_objects as go

# Definisikan warna garis
pick_color = ['#D19C4B', '#FAE8C5', '#FAE8C5', '#FAE8C5']

# Buat grafik line dasar
fig = px.line(
    total_order_per_category,
    x = 'month',
    y = 'total_order',
    color = 'category',
    color_discrete_sequence = pick_color
)

# Definisikan tebal garis
tebal_garis = [8, 5, 5, 5]

# Penebalan Garis yang ingin di highlight
for trace, width in zip(fig.data, tebal_garis):
    if isinstance(trace, go.Scatter):
        trace.line.width = width

# Update layout : Atur ukuran figure dan menambah judul grafik
fig.update_layout(
    height = 600,
    width = 1200,
    title = dict(
        text = "<b>Nostalgia Pizza Classic : 'Old but Gold'</b><br><sup><sup><i>Pizza Classic</i> Mendominasi Penjualan di Tahun 2015<sup><sup>",
        font = dict(
            family = 'sans serif',
            size = 30
        ),
        y = 0.95
    )
)

# Tampilkan Grafik
fig.show()
````

## üë®‚Äçüîß Reparasi Data Viz - 3A : *Decluttering 1*

---

Proses *decluttering* dalam visualisasi data mengacu pada upaya untuk membersihkan dan menyederhanakan visualisasi data agar lebih efektif dan mudah dipahami. Pada grafik yang telah dibuat akan dilakukan proses decluttering sebagai berikut :     

1.   Menghapus *grid* pada grafik
2.   Menghapus label pada sumbu-x karena sudah jelas bahwa axis ticks yang ditampilkan menunjukan kategori nama bulan
3.   Pada sumbu-y perlu dipastikan tidak dimulai dari 0 karena tidak terlalu perlu dan dikhawatirkan trend malah tidak terlihat
4.   Menghapus legend dan menggantinya dengan keterangan di akhir grafik


````python
# Import library yang dibutuhkan
import plotly.express as px
import plotly.graph_objects as go

# Definisikan warna garis
pick_color = ['#D19C4B', '#FAE8C5', '#FAE8C5', '#FAE8C5']

# Buat grafik line dasar
fig = px.line(
    total_order_per_category,
    x = 'month',
    y = 'total_order',
    color = 'category',
    color_discrete_sequence = pick_color
)

# Definisikan tebal garis
tebal_garis = [8, 5, 5, 5]

# Penebalan Garis yang ingin di highlight
for trace, width in zip(fig.data, tebal_garis):
    if isinstance(trace, go.Scatter):
        trace.line.width = width

"""
  Decluttering :
  - Menghapus grid
  - Menghapus label pada sumbu-x
"""
fig.update_layout(
    height = 600,
    width = 1200,
    plot_bgcolor = 'rgba(0, 0, 0, 0)',
    yaxis = dict(
        title = 'Total Order',
        showline = True,
        linewidth = 1,
        color = '#E58651',
        linecolor = '#E58651'
    ),
    xaxis = dict(
        title = '',
        showline = True,
        linewidth = 1,
        color = '#E58651',
        linecolor = '#E58651'
    ),
    title = dict(
        text = "<b>Nostalgia <span style='color:#D19C4B'>Pizza Classic</span> : 'Old but Gold'</b><br><sup><sup><i>Pizza Classic</i> Mendominasi Penjualan di Tahun 2015<sup><sup>",
        font = dict(
            family = 'sans serif',
            size = 30,
            color = '#f7bc60'
        ),
        y = 0.95
    )
)

# Tampilkan Grafik
fig.show()
````

## üë®‚Äçüîß Reparasi Data Viz - 3B : *Decluttering 2* (Mengganti Legend)

Legend dalam grafik atau keterangan tambahan yang menjelaskan unsur-unsur pada grafik yang pada kasus ini menampilkan korelasi warna dan kategori pizza dapat diganti dengan memberikan label langsung pada akhir grafik. Dengan melakukan hal tersebut *audience* dapat langsung mengerti kategori data pada grafik yang ditunjukan.

Untuk dapat melakukan hal tersebut perlu ditandai terlebih dahulu data paling ujung

````python
# Filter data akhir saja (bulan desember)
end_of_data = total_order_per_category[total_order_per_category['month'] == 'December']

# Tampilkan hasil
display(end_of_data)
````

Setelah diperoleh data paling ujung grafik lalu tambahkan titik pada setiap ujung grafiknya

````python
# Import library yang dibutuhkan
import plotly.express as px
import plotly.graph_objects as go

# Definisikan warna garis
pick_color = ['#D19C4B', '#FAE8C5', '#FAE8C5', '#FAE8C5']

# Buat grafik line dasar
fig = px.line(
    total_order_per_category,
    x = 'month',
    y = 'total_order',
    color = 'category',
    color_discrete_sequence = pick_color
)

# Definisikan tebal garis
tebal_garis = [8, 5, 5, 5]

# Penebalan Garis yang ingin di highlight
for trace, width in zip(fig.data, tebal_garis):
    if isinstance(trace, go.Scatter):
        trace.line.width = width

# Menambahkan titik diujung grafik
fig.add_trace(
    go.Scatter(
        x = end_of_data['month'],
        y = end_of_data['total_order'],
        mode = 'markers',
        marker = dict(
            size = [22, 15, 15, 15],
            color = pick_color,
            opacity = 1,
        )
    )
)

"""
  Decluttering :
  - Menghapus grid
  - Menghapus label pada sumbu-x
  - Menghapus legend
"""
fig.update_layout(
    height = 600,
    width = 1200,
    showlegend = False,
    plot_bgcolor = 'rgba(0, 0, 0, 0)',
    yaxis = dict(
        title = 'Total Order',
        showline = True,
        linewidth = 1,
        color = '#E58651',
        linecolor = '#E58651'
    ),
    xaxis = dict(
        title = '',
        showline = True,
        linewidth = 1,
        color = '#E58651',
        linecolor = '#E58651'
    ),
    title = dict(
        text = "<b>Nostalgia <span style='color:#D19C4B'>Pizza Classic</span> : 'Old but Gold'</b><br><sup><sup><i>Pizza Classic</i> Mendominasi Penjualan di Tahun 2015<sup><sup>",
        font = dict(
            family = 'sans serif',
            size = 30,
            color = '#f7bc60'
        ),
        y = 0.95
    )
)

# Tampilkan Grafik
fig.show()
````

Tambahkan perhitungan rata-rata pizza yang diorder berdasarkan kategori pizza sebagai informasi tambahan

````python
# Tambahkan perhitungan rata-rata pizza yang diorder berdasarkan kategori pizza
end_of_data['avg_total_order'] = total_order_per_category.groupby(['category', 'year'])['total_order'].transform(pd.Series.mean)\
                                                         .round(0).astype(int)

# Tampilkan hasil
display(end_of_data)
````

Aplikasikan ke grafik

````python
# Import library yang dibutuhkan
import plotly.express as px
import plotly.graph_objects as go

# Definisikan warna garis
pick_color = ['#D19C4B', '#FAE8C5', '#FAE8C5', '#FAE8C5']

# Buat grafik line dasar
fig = px.line(
    total_order_per_category,
    x = 'month',
    y = 'total_order',
    color = 'category',
    color_discrete_sequence = pick_color
)

# Definisikan tebal garis
tebal_garis = [8, 5, 5, 5]

# Penebalan Garis yang ingin di highlight
for trace, width in zip(fig.data, tebal_garis):
    if isinstance(trace, go.Scatter):
        trace.line.width = width

# Menambahkan titik diujung grafik
fig.add_trace(
    go.Scatter(
        x = end_of_data['month'],
        y = end_of_data['total_order'],
        mode = 'markers',
        marker = dict(
            size = [22, 15, 15, 15],
            color = pick_color,
            opacity = 1,
        )
    )
)

# Tambahkan annotation
for i in end_of_data['ranking']:
    if(i == end_of_data['ranking'].min()):
        teks = f"<b>Pizza {end_of_data[end_of_data['ranking'] == i]['category'].values[0]}</b><br><sup>Avg. Order = {end_of_data[end_of_data['ranking'] == i]['avg_total_order'].values[0]}</sup>"
    else:
        teks = f"<b>Pizza {end_of_data[end_of_data['ranking'] == i]['category'].values[0]}</b><br>"
    fig.add_annotation(
        x = end_of_data[end_of_data['ranking'] == i]['month'].values[0],
        y = end_of_data[end_of_data['ranking'] == i]['total_order'].values[0],
        text = teks,
        showarrow = False,
        font = dict(
            size = 17,
            color = pick_color[i-1],
            family = "sans serif"
        ),
        align = "left",
        xshift = 80,
        yshift = -1,
        opacity = 0.8
    )

"""
  Decluttering :
  - Menghapus grid
  - Menghapus label pada sumbu-x
  - Menghapus legend
"""
fig.update_layout(
    height = 600,
    width = 1200,
    showlegend = False,
    plot_bgcolor = 'rgba(0, 0, 0, 0)',
    yaxis = dict(
        title = 'Total Order',
        showline = True,
        linewidth = 1,
        color = '#E58651',
        linecolor = '#E58651'
    ),
    xaxis = dict(
        title = '',
        showline = True,
        linewidth = 1,
        color = '#E58651',
        linecolor = '#E58651'
    ),
    title = dict(
        text = "<b>Nostalgia <span style='color:#D19C4B'>Pizza Classic</span> : 'Old but Gold'</b><br><sup><sup><i>Pizza Classic</i> Mendominasi Penjualan di Tahun 2015<sup><sup>",
        font = dict(
            family = 'sans serif',
            size = 30,
            color = '#F7BC60'
        ),
        y = 0.95
    )
)

# Tampilkan Grafik
fig.show()
````

## üë®‚Äçüîß Reparasi Data Viz - 4 : *Hover to Highlight*

Saat kursor diarahkan ke titik data maka secara otomatis akan ditampilkan informasi mengenai kategori pizza, bulan dan total ordernya. Tentu dapat ditambahkan informasi lainnya jika diperlukan.

docs : <i>https://plotly.com/python/hover-text-and-formatting/</i>

````python
# Import library yang dibutuhkan
import plotly.express as px
import plotly.graph_objects as go

# Definisikan warna garis
pick_color = ['#D19C4B', '#FAE8C5', '#FAE8C5', '#FAE8C5']

# Buat grafik line dasar
fig = px.line(
    total_order_per_category,
    x = 'month',
    y = 'total_order',
    color = 'category',
    color_discrete_sequence = pick_color,
    hover_data = ['category', 'year']
)

# Edit Hover Template
fig.update_traces(
    hovertemplate = 'Pizza %{customdata[0]}<br>%{x} %{customdata[1]}<br><sup>Total Order = %{y}</sup>',
    hoverlabel = dict(
        bgcolor = 'white',
        font = dict(
            size = 14,
            color = '#D19C4B'
        )
    )
)

# Definisikan tebal garis
tebal_garis = [8, 5, 5, 5]

# Penebalan Garis yang ingin di highlight
for trace, width in zip(fig.data, tebal_garis):
    if isinstance(trace, go.Scatter):
        trace.line.width = width

# Menambahkan titik diujung grafik
fig.add_trace(
    go.Scatter(
        x = end_of_data['month'],
        y = end_of_data['total_order'],
        mode = 'markers',
        marker = dict(
            size = [22, 15, 15, 15],
            color = pick_color,
            opacity = 1,
        ),
        text = end_of_data['category'],
        hovertemplate = "Pizza %{text}<br>%{x} 2015<br><sup>Total Order = %{y}</sup>"
    )
)

# Tambahkan annotation
for i in end_of_data['ranking']:
    if(i == end_of_data['ranking'].min()):
        teks = f"<b>Pizza {end_of_data[end_of_data['ranking'] == i]['category'].values[0]}</b><br><sup>Avg. Order = {end_of_data[end_of_data['ranking'] == i]['avg_total_order'].values[0]}</sup>"
    else:
        teks = f"<b>Pizza {end_of_data[end_of_data['ranking'] == i]['category'].values[0]}</b><br>"
    fig.add_annotation(
        x = end_of_data[end_of_data['ranking'] == i]['month'].values[0],
        y = end_of_data[end_of_data['ranking'] == i]['total_order'].values[0],
        text = teks,
        showarrow = False,
        font = dict(
            size = 17,
            color = pick_color[i-1],
            family = "sans serif"
        ),
        align = "left",
        xshift = 80,
        yshift = -1,
        opacity = 0.8
    )

"""
  Decluttering :
  - Menghapus grid
  - Menghapus label pada sumbu-x
  - Menghapus legend
"""
fig.update_layout(
    height = 600,
    width = 1200,
    showlegend = False,
    plot_bgcolor = 'rgba(0, 0, 0, 0)',
    yaxis = dict(
        title = 'Total Order',
        showline = True,
        linewidth = 1,
        color = '#E58651',
        linecolor = '#E58651'
    ),
    xaxis = dict(
        title = '',
        showline = True,
        linewidth = 1,
        color = '#E58651',
        linecolor = '#E58651'
    ),
    title = dict(
        text = "<b>Nostalgia <span style='color:#D19C4B'>Pizza Classic</span> : 'Old but Gold'</b><br><sup><sup><i>Pizza Classic</i> Mendominasi Penjualan di Tahun 2015<sup><sup>",
        font = dict(
            family = 'sans serif',
            size = 30,
            color = '#F7BC60'
        ),
        y = 0.95
    )
)

# Tampilkan Grafik
fig.show()
````
